### Loading the necessary libraries.

```python

import pandas as pd 
import numpy as np 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report 
from sklearn.svm import SVC
```

### Reading training Data set 

```python
df1=pd.read_csv(r'C:\Users\DAKSHIN\Desktop\hack\train_MpHjUjU.csv')
```

### Dropping all columns with Dates


```python
df1=df1.drop('MMM-YY',axis=1)
df1=df1.drop('Dateofjoining',axis=1)
df1=df1.drop('LastWorkingDate',axis=1)
```

### Grouping all coulumns and summing Total Business value to reduce duplicates and concatinating into a new data frame 

```python
g=df1.groupby('Emp_ID')['Emp_ID','Age','Gender','City','Education_Level','Salary','Joining Designation','Designation',
                       'Quarterly Rating','Target'].tail(1)
g.reset_index(drop=True,inplace=True)
gg=df1.groupby('Emp_ID').agg({'Total Business Value':'sum'})
gg.reset_index(drop=True,inplace=True)
df2=pd.concat([g,gg],axis=1,ignore_index=False)
```

```python
df2.head(10)
```

### Scaling Numerical Columns

```python

df2[['Salary', 'Total Business Value','Age']] = StandardScaler().fit_transform(df2[['Salary', 'Total Business Value','Age']])
df2
```

### Assigning Label Encoder funtion to variable 

```python

le=LabelEncoder()
```

### Label Encoding the Categorical Variables

```python
df2['Gender']=le.fit_transform(df2['Gender'])
df2['Education_Level']=le.fit_transform(df2['Education_Level'])
df2['City']=le.fit_transform(df2['City'])
```

```python
df2
```

### Dropping null values

```python
df2=df2.dropna()
```

```python
df2=df2.drop('Emp_ID',axis=1)
```

### Checking Length and Up sampling 

```python
yes = df2[df2.Target==0] 
no = df2[df2.Target==1] 
print(len(yes))
print(len(no))
```

```python
ones=df2[df2.Target==1]
zeros=df2[df2.Target==0]
up_zeros=resample(zeros,replace=True,n_samples=len(ones),random_state=147)
df3=pd.concat([ones,up_zeros])
```

```python
df3
```

### Assigning Independent Variables and Dependent Variable

```python
x=df3.drop('Target',axis=1)
y=df3['Target']
```

### Training Test Split


```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=121)
```

### SVC Model Building

```python
svc = SVC(kernel='linear', probability=True, C=1, gamma='auto')
svc_model = svc.fit(xtrain, ytrain)
ypred=svc_model.predict(xtest)
```

```python
print(classification_report(ytest,ypred))
```
